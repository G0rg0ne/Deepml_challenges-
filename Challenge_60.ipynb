```python
import numpy as np
import math
```

```python


def compute_tf_idf(corpus, query):
    final_result = np.zeros((len(corpus),len(query)))
    N = len(corpus)
    for term in query:
        df = sum(term in doc for doc in corpus)
        idf = math.log((1 + N) / (1 + df)) + 1
        for doc in corpus:
            tf = doc.count(term) / len(doc)
            tf_idf = round( tf * idf, 5)
            final_result[corpus.index(doc),query.index(term)] = tf_idf

    return final_result

	
```

```python
import numpy as np 
corpus = [ ["the", "cat", "sat", "on", "the", "mat"], ["the", "dog", "chased", "the", "cat"], ["the", "bird", "flew", "over", "the", "mat"] ] 
query = ["cat", "mat"] 
print(compute_tf_idf(corpus, query))
```
