```python
import numpy as np
```

```python

def accuracy_score(y_true, y_pred):
    correct = (y_true == y_pred).sum()
    accuracy = correct / len(y_true)
    return accuracy 
```

```python
y_true = np.array([1, 0, 1, 1, 0, 1])
y_pred = np.array([1, 0, 0, 1, 0, 1])
output = accuracy_score(y_true, y_pred)
print(output)
```
