```python
num_ex = [12, 13, 14]
```

```python
def ham(nums):
   total = 0
   n = len(nums)
   for i in range(32):
    ones = 0
    for num in nums:
        ones += (num>>i) & 1
    zeros = n - ones
    total += ones * zeros
   return total

ham(num_ex)
```
