```python
import numpy as np
```

```python

def translate_object(points, tx, ty):
    translated_points = []
    for i in range(len(points)):
        translated_points+= [[float(points[i][0]+tx), float(points[i][1]+ty)]]
    return translated_points
```

```python
points = [[0, 0], [1, 0], [0.5, 1]]
tx, ty = 2, 3       

print(translate_object(points, tx, ty))     
```
