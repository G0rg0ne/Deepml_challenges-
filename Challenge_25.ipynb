```python
import numpy as np
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
    
def mean_squared_error(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)

def train_neuron(features, labels, weights, bias, learning_rate, epochs):
    features = np.asarray(features)
    labels = np.asarray(labels)
    weights = np.asarray(weights)

    for epoch in range(epochs):
        z = np.dot(features, weights) + bias
        a = sigmoid(z)

        N = features.shape[0]
        a_gradient = (a - labels) * a * (1 - a) / N

        gradient_weights = np.dot(features.T, a_gradient)
        gradient_bias = np.sum(a_gradient)

        weights -= learning_rate * gradient_weights
        bias -= learning_rate * gradient_bias
        
        mse = mean_squared_error(labels, a)

        print(mse)

    return weights, bias, mse
```

```python


def sigmoid(z: np.ndarray) -> np.ndarray:
    return 1 / (1 + np.exp(-z))


def train_neuron(
    features: np.ndarray,
    labels: np.ndarray,
    initial_weights: np.ndarray,
    initial_bias: float,
    learning_rate: float,
    epochs: int
):

    X = np.asarray(features, dtype=float)
    y = np.asarray(labels, dtype=float)
    weights = np.asarray(initial_weights, dtype=float)
    bias = float(initial_bias)

    mse_values = []
    N = X.shape[0]

    for _ in range(epochs):

        z = np.dot(X, weights) + bias
        a = sigmoid(z)


        mse = np.mean((a - y) ** 2)
        mse_values.append(round(mse, 4))

        dz = 2 * (a - y) * a * (1 - a) / N

        grad_w = np.dot(X.T, dz)
        grad_b = np.sum(dz)

  
        weights -= learning_rate * grad_w
        bias -= learning_rate * grad_b

    return weights.round(4).tolist(), round(bias, 4), mse_values
```

```python

features = [[1.0, 2.0], [2.0, 1.0], [-1.0, -2.0]]
labels = [1, 0, 0]
initial_weights = [0.1, -0.2]
initial_bias = 0.0
learning_rate = 0.1
epochs = 10
train_neuron(features, labels, initial_weights, initial_bias, learning_rate, epochs)
```
