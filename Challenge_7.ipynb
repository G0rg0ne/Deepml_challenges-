```python
import numpy as np
```

```python

def is_invertible(matrix): 
    det = matrix[0][0]*matrix[1][1] - matrix[0][1]*matrix[1][0]
    if det != 0:
        return True
    else :
        return False

def transform_matrix(A: list[list[int|float]], T: list[list[int|float]], S: list[list[int|float]]) -> list[list[int|float]]:
    try:
        A = np.array(A)
        T = np.array(T)
        S = np.array(S)
        T_invrs = np.linalg.inv(T)

        transformed_matrix = T_invrs @ A @ S
        return transformed_matrix
    except np.linalg.LinAlgError:
        return -1

```

```python
A = [[1, 2], [3, 4]]
T = [[2, 0], [0, 2]]
S = [[1, 1], [0, 1]]
transform_matrix(A,T,S)
```

```python
print(transform_matrix([[1, 2, 3],[0, 1, 4],[5, 6, 0]], [[2, 0, 0],[0, 2, 0],[0, 0, 2]], [[0, 1, 0],[0, 0, 1],[1, 0, 0]]))
```
