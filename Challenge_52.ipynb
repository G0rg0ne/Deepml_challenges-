```python
import numpy as np
```

```python
def recall(y_true, y_pred):
    tp = np.sum(y_true * y_pred)
    positives = np.sum(y_true)
    return tp / positives
```

```python
import numpy as np

y_true = np.array([1, 0, 1, 1, 0, 1])
y_pred = np.array([1, 0, 1, 0, 0, 1])

print(recall(y_true, y_pred))
```
