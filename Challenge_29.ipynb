```python
import numpy as np

def shuffle_data(X, y, seed=None):
    """
    Shuffle X and y using NumPy's legacy RNG
    to match deterministic expected output.
    """
    n_samples = X.shape[0]
    indices = np.arange(n_samples)

    if seed is not None:
        np.random.seed(seed)  # legacy RNG seed

    np.random.shuffle(indices)  # legacy RNG shuffle

    return X[indices], y[indices]
```
