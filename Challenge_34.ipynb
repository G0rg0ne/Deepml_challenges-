```python
import numpy as np
```

```python

def to_categorical(x, n_col=None):
    unique_values = np.unique(x)
    dict_code = {}
    for i in range(len(unique_values)):
       code= np.zeros(n_col if n_col else len(unique_values))
       code[unique_values[i]] = 1
       dict_code[unique_values[i]]= code
    result = []
    for e in x:
        result.append(dict_code[e])
    return result
```

```python
x = np.array([0, 1, 2, 1, 0])
output = to_categorical(x)
print(output)
```
